### Return on Investment (ROI): Total Tution (4-years) vs. Starting Salary (first 4-years)

```python
#GIVE VALUES TO COLUMNS TO MANIPULATE
starting_salary = combined_merged_data["Starting Median Salary"]
starting_salary_FourYrs = starting_salary*20
cost_2015 = combined_merged_data["Cost 2015"]
cost_2014 = combined_merged_data["Cost 2014"]
school_name = combined_merged_data["School Name"]
per_change = combined_merged_data["% Change"]
state = combined_merged_data["State"]
region = combined_merged_data["Region"]
total_tuition = (cost_2014+cost_2015)*2
mid_salary = combined_merged_data["Mid-Career Median Salary"]
#EARNINGS POTENTIAL
#roi = NET PROFIT (income) / TOTAL INVESTMENT (invested capital)x100
#Starting salary ~0-5 years
roi = (total_tuition/starting_salary_FourYrs)*100
roi2 = (total_tuition/mid_salary)*100
```

```python
#ADD NEW COLUMNS
combined_merged_data["ROI"] = roi
combined_merged_data["ROI (Mid-Career)"] = roi2
combined_merged_data["Undergrad Tuition (4-Years)"] = total_tuition
```

## OBSVERVATION:
There are colleges that have an outstanding ROI when comparing the first four years in an entry-level (~3 years) position and the cost of a undergraduate degree (e.g. MIT, Georgia Tech, Brown University, etc.). 

*The ROIs are ONLY based on starting median salary (20 years) & cost of attendance (4 years)

```python
roi_salaries_df = combined_merged_data[["School Name","State","Region","Starting Median Salary",\
                                       "Undergrad Tuition (4-Years)", "ROI"]]
#,"% Change","Mid-Career Median Salary","ROI (Mid-Career)"]]
#Sorting from highest to lowest 'difference'
roi_highest = roi_salaries_df.sort_values("ROI",ascending=False)
roi_highest.head(10)
```

## OBSVERVATION:
The universities with the lowest ROI include top universities such as Stanford, Rice, and Duke. Thus, leading us to the conclusion that ivy league education doesn't necessarily mean you'll be earning higher wages. 

*The ROIs are ONLY based on starting median salary (20 years) & cost of attendance (4 years)

```python
#Sorting from highest to lowest 'difference'
roi_lowest = roi_salaries_df.sort_values("ROI",ascending=True)
roi_lowest.head(10)
```
